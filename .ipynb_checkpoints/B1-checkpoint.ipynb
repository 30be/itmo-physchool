```python
from math import *
from matplotlib import pyplot as plt
from matplotlib.patches import Circle

def count_collisions_until_hit(a, b, l, sphere_x, sphere_y, sphere_r, angle):
    intersect_sphere = lambda x, y, angle : abs(angle - atan2(y, x)) < asin(sphere_r / sqrt(x**2+y**2))
    walls = lambda x, y, angle : [(-x / cos(angle),      0,                        y - x * tan(angle),       pi - angle), # left
                                  ((a - x) / cos(angle), a,                        y + (a - x) * tan(angle), pi - angle), # right
                                  (-y  / sin(angle),     x - y / tan(angle),       0,                        -angle),     # bottom
                                  ((b - y) / sin(angle), x + (b - y) / tan(angle), b,                        -angle)]     # top
    collide = lambda *state : min(walls(*state),  key = lambda c : c[0] if c[0] > .001 else float("inf"))[1:]   # Find minimal positive distance
    yield (state := (l, 0, angle * pi / 180))
    for i in range(10000):
        if not intersect_sphere(sphere_x - state[0], sphere_y - state[1], state[2]):
            yield (state := collide(*state))
    yield (state := collide(*state))
```

```python
path = list(count_collisions_until_hit(a:=7, b:=5, l:=2, sphere_x:=6, sphere_y:=4, sphere_r:=.4, angle:=10))
print("Collisions made:", len(path) - 2)
```

```python
plt.gca().add_patch(Circle((sphere_x, sphere_y), sphere_r, color='r'))
for (x1, y1, _), (x2, y2, _) in zip(path, path[1:]):
    plt.plot([x1, x2], [y1, y2], 'b')
plt.margins(0, 0)
```
