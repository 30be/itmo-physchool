```python
from math import *

MAX = 10000

def count_collisions_until_hit(a, b, l, sphere_x, sphere_y, sphere_r, angle):
    def intersect_sphere(x, y, angle):
        dx = sphere_x - x
        dy = sphere_y - y
        angle_to_sphere = atan2(dy, dx)
        angular_diameter = asin(sphere_r / sqrt(dx**2+dy**2))
        return abs(angle - angle_to_sphere) < angular_diameter
        
    def collide(x, y, angle):
        cos_a, sin_a = cos(angle), sin(angle)
        wall, t = min([
                ("left",       -x  / cos_a),
                ("right",  (a - x) / cos_a),
                ("bottom",     -y  / sin_a),
                ("top",    (b - y) / sin_a),
            ],  key = lambda c : c[1] if c[1] > 0.001 else float("inf")) 
        new_x = 0 if wall == 'left' else a if wall == 'right' else x + t * cos_a
        new_y = 0 if wall == 'bottom' else b if wall == 'top' else y + t * sin_a
        new_angle = -angle + pi if wall in ['left', 'right'] else -angle
        return new_x, new_y, new_angle

    state = (l, 0, angle * pi / 180)
    path = []
    for i in range(MAX):
        if intersect_sphere(*state):
            path.append(state)
            path.append(collide(*state))
            return path
        path.append(state)
        state = collide(*state)
    return path
```

```python
a, b = 7, 5
l = 2
sphere_x, sphere_y, sphere_r = 6, 4, 0.4
angle = 10
path = count_collisions_until_hit(a,b,l, sphere_x, sphere_y, sphere_r, angle)
print("Collisions made:", len(path)-2)
```

```python
from matplotlib import pyplot as plt
from matplotlib.patches import Circle
fig, ax = plt.subplots()
ax.plot([0, a, a, 0, 0], [0, 0, b, b, 0], 'k-')
circle = Circle((sphere_x, sphere_y), sphere_r, color='r')
ax.add_patch(circle)

for i in range(len(path) - 1):
    x1, y1, _ = path[i]
    x2, y2, _ = path[i+1]
    ax.plot([x1, x2], [y1, y2], 'b')
```
