```python
import numpy as np
import matplotlib.pyplot as plt

m = 0.1
k = 4
b = 0.2
x0 = 0.05
v0 = 0

gamma = b / (2 * m)
omega_0 = np.sqrt(k/m - gamma**2)
A = np.sqrt(x0**2 * omega_0**2 + (v0 + gamma * x0)**2) / omega_0
phi = np.arctan2(-(v0 + gamma * x0), x0 * omega_0)

def x(t):
    return A * np.exp(-gamma * t) * np.cos(omega_0 * t + phi)
def v(t):
     return A * np.exp(-gamma * t) * (-gamma * np.cos(omega_0 * t + phi) - omega_0 * np.sin(omega_0 * t + phi))

t_values = np.linspace(0, 5, 500)

plt.plot(t_values, x(t_values))
plt.title('–ó–∞–≤–∏—Å–∏–º–æ—Å—Ç—å x(t) –¥–ª—è –∑–∞—Ç—É—Ö–∞—é—â–∏—Ö –∫–æ–ª–µ–±–∞–Ω–∏–π')
plt.xlabel('t, —Å')
plt.ylabel('x, –º')
plt.grid(True)

T_theoretical = 2 * np.pi / omega_0
print(f"T = 2*pi / omega_0 = {T_theoretical:.4f} —Å")
```

```python
from scipy.integrate import quad

def integrand_energy(t):
    return b * (v(t))**2
dissipated_energy, error = quad(integrand_energy, 0, 2)

print(dissipated_energy)
```

```python
import itertools
import matplotlib.pyplot as plt 

m = 0.1
k = 4
b = 0.2
x0 = 0
v0 = 5
d = 0.02
omega_0_sq = k / m
gamma = b / (2 * m)
dt = 0.001
steps = int(5 / dt)

def advance(x, v):
    dx = v * dt
    dv = (-(k/m) * x - (b/m) * v) * dt
    if (x > d):
        return (d, -v)
    return (x + dx, v + dv)

states = list(itertools.accumulate([(x0,v0)] + [None] * (steps - 1), lambda prev_state, _: advance(*prev_state)))
xs = [item[0] for item in states]

collisions = [x for x in xs if x > d]
plt.title('–î–≤–∏–∂–µ–Ω–∏–µ –∑–∞—Ç—É—Ö–∞—é—â–µ–≥–æ –æ—Å—Ü–∏–ª–ª—è—Ç–æ—Ä–∞ —Å–æ —Å—Ç–µ–Ω–∫–æ–π')
plt.xlabel('t, —Å')
plt.ylabel('x, –º')
plt.grid(True)
plt.plot([i * dt for i in range(steps)], xs)

print(len(collisions), "collisions overall")
```
