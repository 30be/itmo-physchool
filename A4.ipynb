```python
import requests
import math
data = requests.get("https://cloud.physics.itmo.ru/s/pY68e5CrdtQHF6M/download").text.replace(',', '.').splitlines()
wavelengths = [float(line.split('\t')[0]) for line in data]
values      = [float(line.split('\t')[1]) for line in data]
```

```python
average = sum(values) / len(values) # Среднее величины на измеряемом промежутке
mean = sum(v * w for w, v in zip(wavelengths, values)) / sum(values) # Взвешенное среднее
variance = sum([v * (w - mean)**2 for w, v in zip(wavelengths, values)]) / sum(values) # Дисперсия
sd = math.sqrt(variance) # Среднеквадратичное отклонение
print("average:", average)
print("mean:", mean)
print("standard deviation:", sd)
```
